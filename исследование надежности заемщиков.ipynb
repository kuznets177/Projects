```python
total_income_mean = df['total_income'].mean()
df['total_income'] = df['total_income'].fillna(value=total_income_mean)

#zero = df['days_employed'][df['days_employed']> 0] #сортируем значения days_mployed >0
#df['days_employed'] = (df['days_employed']/24)/365   #переводим рабочие часы в дни
def days_true(row):
    age = row['dob_years']
    days = row['days_employed']
    if 18 <= age <= 65:
        days = ((age - 18) * 248) # беру трудоспособных клиентов, вычесть 18 лет и умножить на среднее число рабочих дней в году 248
    return days
 
df['days_employed'] = df.apply(days_true, axis = 1)

days_employed_median = df['days_employed'].median() 
df['days_employed'] = df['days_employed'].fillna(value=days_employed_median)
df['days_employed'] = df['days_employed'].apply(abs) # переводим отрицательные значения в положительные
df.isnull().sum()



#df['dob_years'] = df.sort_values(by='dob_years', ascending=True) #сортируем столбец возраст, чтобы соотнести стаж и возраст
#df['days_employed']
#df.info()
#df['zero'] = df['days_employed'][df['days_employed']> 0] #сортируем значения days_mployed >0







```

```python
# Разбиваем purpose на категории по леммам: 'Жилье/недвижимость', 'автомобиль', 'образование', 'свадьба'
def lemming(row):
    lema = ' '.join(m.lemmatize(row))
        
    if 'автом' in lema:
        return 'автомобиль'
    if 'свадьб' in lema:
        return 'свадьба'
    if 'образов' in lema: 
        return 'образование'
    if 'недвиж' or 'жиль' in lema:
        return 'недвижимость'
    else:
        return purpose
        
df['purpose_lem'] = df['purpose'].apply(lemming)
df
```

Здесь мы лемматизмруем цели кредитов и приводим их к типу строк методом join(), т.к. со строками работать проще, чем со списками лемм.
После этого выделяем уникальные цели кредитов. Здесь я объединил цели жилье и недвижимость и свел их в общую категорию - недвижимость, поскольку это одна и так же категория, просто формулирвка разная. Делим цели на большие наиболее подходящие группы (недвижимость, автомобиль, образование и свадьба).
