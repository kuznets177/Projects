```python
# пишем функцию, которая будет создавать сводную и выводить график или ящик у усами в зависимости от здачи

def pivot_plt(data, index, values, aggfunc, column, color, title, y_unit, x_unit):
    
    (data
     .pivot_table(index=index,
                      values=[values],
                      aggfunc=aggfunc)
     #.sort_values(by=column, ascending=True)
     .plot(kind='bar',figsize=(15, 6), fontsize=15, color='g')
    )   
    plt.title(title)
    plt.ylabel(y_unit)
    plt.xlabel(x_unit)
    plt.legend(values)
    plt.show()
    plt.close()
    
    
```

```python
# Посмотрите, сколько игр выпускалось в разные годы. 
pivot_plt(df, 'year_of_release', 'name', 'count', 'year_of_release', 'green', 'Кол-во релизов игр по годам', 'Кол-во игр', 'Год релиза')
```

<div class="alert alert-success">
<h2> Комментарий от ревьюера</h2>

Отлично. Согласна с определенным временем жизни платформ.

```python
# Найдем объем продаж для каждой платформы по годам, для этого создадим сводную
sales_volume = (df
                .pivot_table(
                index='year_of_release',
                values='total_sales',
                columns='platform',
                aggfunc='sum')
                .sort_values(by='year_of_release'))

sales_volume.fillna(0, inplace=True)
np.ceil(sales_volume).astype('Int64')

# подтвердим наши данные расчетами
# соберем список из количества лет жизни каждой платформы, и уберем оттуда нули
notnull=[]
for platform in sales_volume.columns:
    (notnull
     .append(sales_volume[platform]
             .ne(0)
             .value_counts()
             .loc[sales_volume[platform]
                  .ne(0)
                  .value_counts()
                  .index == True]
             .values[0]))
    

# построим ящик с усами для продолжительности жизни игровых платформ,чтобы подтвердить наши данные из графика выше
plt.figure(figsize=(15, 3))
sns.boxplot(data=notnull, orient='h')
plt.suptitle('Средняя продолжительности жизни игровых платформ')
_ = plt.ylim()
```

Таким образом мы точно можем сказать, что в среднем, платормы существуют около 8 лет. Но наше среднее рассчитано с учетом данных с 1980-х годов, когда технологии были еще слабо развиты. Сейчас раззитие идет намного быстрее и мы понимаем, что раасматривать данные за весь период не корректно, поскольку с каждым годом темп развития технологий растет, а все устаревшие данные не отностяся к настоящему. Поэтому мы делаем срез данных, в котором убираем данные до 2000 года. Таким образом, исходя из срока жизни платформ мы сможем проанализировать актуальные данные по платформам.

## Шаг 5. Проверьте гипотезы

```python
# импортируем иблиотетку scipy для сравнения дисперсий выборок
from scipy.stats import levene
import scipy.stats as stats

# обозначаем alpha = 0.05
alpha = .05
```

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H0 - Средние пользовательские рейтинги платформ <b>одинаковые</b>., т.е. средние значения равны

H1 - Средние пользовательские рейтинги платформ <b>не одинаковые</b>., т.е. средние значения различаются (не равны)

Для проверки гипотезы применим критерий p-value, т.к. он лучше всего подходит в данной ситуации сравнения по двум платформам.

```python
# смотри средние
xone_pc = df_actual_period.query('platform == "XOne" or platform == "PC"')
xone_pc.pivot_table(index='platform', values='user_score', aggfunc='mean')
```

Средние значения почти на одном уровне

```python
xone = df_actual_period.query('platform == "XOne" and user_score >= 0')['user_score']
pc = df_actual_period.query('platform == "PC" and user_score >= 0')['user_score']


# Проверяем, принадлежат ли xone и pc платформам с равной дисперсией.
stat, p = levene(xone, pc)
print(p)
```

Малое значение p предполагает, что совокупности не имеют одинаковых дисперсий.

```python
# добавляем сравнение дисперсий выборок 
xone_var = [np.var(x, ddof=1) for x in [xone]]
pc_var = [np.var(x, ddof=1) for x in [pc]]

print(f'Дисперсия платформы XOne: {(xone_var)}')
print(f'Дисперсия платформы PC: {(pc_var)}')

# смотрим отношение дисперсий, оно явно больше 5%
print(stats.f_oneway(xone, pc))
```

Как мы видим из данных, разница между дисперсиями значительная.

```python
# т.к.отношение дисперсий больше 5%, то также оставим equal_var = false
results = st.ttest_ind(
    xone, 
    pc, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Статистический тест показал, что гипотезу о равенстве среднего пользовательского рейтинга для XOne и PC не получилось отвергнуть, т.е. p > 0.05. Если бы мы утверждали, что средние рейтинги для этих платформ различаются, мы бы допустили ошибку с вероятностью 20%.

### Средние пользовательские рейтинги жанров Action  и Sports разные

H0 - Средние пользовательские рейтинги жанров Action и Sports <b>одинаковые</b>, т.е. средние значения равны

H1 - Средние пользовательские рейтинги жанров Action и Sports <b>разные</b>, т.е. средние значения различаются (не равны)

Для проверки гипотезы применим критерий p-value, т.к. он так же хорошо подходит в данном случае.

```python
# обозначаем alpha = 0.05
alpha = .05

# смотри средние по выборкам
action_sports = df_actual_period.query('genre == "Action" or genre == "Sports"')
action_sports.pivot_table(index='genre', values='user_score', aggfunc='mean')
```

Средне значения почти на одном уровне

```python
action = df_actual_period.query('genre == "Action" and user_score >= 0')['user_score']
sports = df_actual_period.query('genre == "Sports"  and user_score >= 0')['user_score']

# Проверяем, принадлежат ли xone и pc платформам с равной дисперсией.
stat, p = levene(action, sports)
print(p)
```

Высокое значение p предполагает, что совокупности имеют одинаковые дисперсии.

```python
# добавляем сравнение дисперсий выборок 
action_var = [np.var(x, ddof=1) for x in [action]]
sports_var = [np.var(x, ddof=1) for x in [sports]]

print(f'Дисперсия жанра Action: {(action_var)}')
print(f'Дисперсия жанра Sports: {(sports_var)}')

# смотрим отношение дисперсий, оно явно больше 5%
print(stats.f_oneway(action, sports))
```

```python
results = st.ttest_ind(
    action, 
    sports, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
